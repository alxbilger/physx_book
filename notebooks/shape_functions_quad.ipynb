# Shape Functions on bilinear quadrilateral

This notebook derives the bilinear shape functions for a 2D finite element (quadrilateral).
Shape functions $N_i$ are used to interpolate values within an element based on nodal values.


```python
from sympy import symbols, Eq, solve, simplify, IndexedBase, factor, diff
```

```python
from typst_utils import typst_print
```

## Symbol Definition
We define the 2D position vector $\mathbf{X} = [\xi, \eta]^T$ and the positions of the four nodes $\mathbf{X}_i$ for $i=0, 1, 2, 3$.


```python
# Define symbols
X = IndexedBase('referenceposition')
# Using IndexedBase for each node to represent vectors
nodes = [IndexedBase(f'referenceposition^(({i}))') for i in range(4)]
```

## Bilinear Combination
We assume that each shape function $N_i$ is a bilinear function of the components of $\mathbf{X}$:
$N_i(\mathbf{X}) = a \cdot X[0] \cdot X[1] + b \cdot X[0] + c \cdot X[1] + d$


```python
# Define shape functions as bilinear combinations: Ni = a*X[0]*X[1] + b*X[0] + c*X[1] + d
a, b, c, d = symbols('a b c d')
```

## Solving the System
The shape functions must satisfy the Kronecker delta property: $N_i(\mathbf{X}_j) = \delta_{ij}$.
This leads to a system of four linear equations for the coefficients $a, b, c$ and $d$ for each $N_i$.


```python
# Solve for each shape function N[i]
N = []
for i in range(4):
    # Ni(X) = a*X[0]*X[1] + b*X[0] + c*X[1] + d
    eqs = [
        Eq(a * nodes[j][0] * nodes[j][1] + b * nodes[j][0] + c * nodes[j][1] + d, 1 if i == j else 0)
        for j in range(4)
    ]
    sol = solve(eqs, (a, b, c, d))
    print(f"for N[{i}]:\na={sol[a].simplify()};\nb={sol[b].simplify()};\nc={sol[c].simplify()};\nd={sol[d].simplify()}\n\n")
    Ni = sol[a] * X[0] * X[1] + sol[b] * X[0] + sol[c] * X[1] + sol[d]
    N.append(Ni.simplify())
```

## Symbolic Results
The general symbolic expressions for the shape functions are:


```python
# Print results
for i in range(4):
    print(f"Shape function N[{i}]:\n{N[i]}\n")
```

## Symbolic Gradients
The symbolic gradients of the shape functions with respect to $\xi$ and $\eta$ are:


```python
# Compute gradients
dN_dxi = [diff(Ni, X[0]) for Ni in N]
dN_deta = [diff(Ni, X[1]) for Ni in N]

# Print gradients
for i in range(4):
    print(f"Gradient dN[{i}]/dxi: {dN_dxi[i]}")
    print(f"Gradient dN[{i}]/deta: {dN_deta[i]}")
    print()
```

## Numerical Evaluation
We can specialize these general expressions for specific reference elements.


### Standard Reference Quadrilateral (-1,-1), (1,-1), (1,1), (-1,1)


```python
# Standard reference quad values
vals_std = {
    nodes[0][0]: -1, nodes[0][1]: -1,
    nodes[1][0]: 1, nodes[1][1]: -1,
    nodes[2][0]: 1, nodes[2][1]: 1,
    nodes[3][0]: -1, nodes[3][1]: 1
}

N_numeric_std = [Ni.subs(vals_std) for Ni in N]
dN_dxi_numeric_std = [dNi.subs(vals_std) for dNi in dN_dxi]
dN_deta_numeric_std = [dNi.subs(vals_std) for dNi in dN_deta]


print("Shape functions:")
for i in range(4):
    print(f"{typst_print(factor(N_numeric_std[i]))};")

print("\nGradients:")
for i in range(4):
    print(f"{typst_print(dN_dxi_numeric_std[i])},")
    print(f"{typst_print(dN_deta_numeric_std[i])};")
    print()
```

### Standard Reference Quadrilateral (0,0), (1,0), (0,1), (1,1)


```python
# Standard reference quad values
vals_std = {
    nodes[0][0]: 0, nodes[0][1]: 0,
    nodes[1][0]: 1, nodes[1][1]: 0,
    nodes[2][0]: 0, nodes[2][1]: 1,
    nodes[3][0]: 1, nodes[3][1]: 1
}

N_numeric_std = [Ni.subs(vals_std) for Ni in N]
dN_dxi_numeric_std = [dNi.subs(vals_std) for dNi in dN_dxi]
dN_deta_numeric_std = [dNi.subs(vals_std) for dNi in dN_deta]

print("Shape functions:")
for i in range(4):
    print(f"{typst_print(factor(N_numeric_std[i]))};")

print("\nGradients:")
for i in range(4):
    print(f"{typst_print(dN_dxi_numeric_std[i])},")
    print(f"{typst_print(dN_deta_numeric_std[i])};")
    print()
```
