# Shape Functions on linear edge

This notebook derives the linear shape functions for a 1D finite element (edge).
Shape functions $N_i$ are used to interpolate values within an element based on nodal values.


```python
from sympy import symbols, Eq, solve, diff, IndexedBase
```

```python
from typst_utils import typst_print
```

## Symbol Definition
We define the natural coordinate $\xi$ and the positions of the two nodes $\xi_0$ and $\xi_1$.


```python
# Define symbols
xi = symbols('referenceposition')
nodes = [IndexedBase(f'referenceposition^(({i}))') for i in range(2)]
```

## Linear Combination
We assume that each shape function $N_i$ is a linear function of $\xi$:
$N_i(\xi) = a \cdot \xi + b$


```python
# Define shape functions as linear combinations: Ni = a * xi + b
a, b = symbols('a b')
```

## Solving the System
The shape functions must satisfy the Kronecker delta property: $N_i(\xi_j) = \delta_{ij}$.
This leads to a system of two linear equations for the coefficients $a$ and $b$ for each $N_i$.


```python
# Solve for each shape function N[i]
N = []
for i in range(2):
    # Ni(xi) = a*xi + b
    eqs = [
        Eq(a * nodes[0] + b, 1 if i == 0 else 0),
        Eq(a * nodes[1] + b, 1 if i == 1 else 0)
    ]
    sol = solve(eqs, (a, b))
    Ni = sol[a] * xi + sol[b]
    N.append(Ni.simplify())
```

## Symbolic Results
The general symbolic expressions for the shape functions are:


```python
# Print results
for i in range(2):
    print(f"{typst_print(N[i])};")
```

## Symbolic Gradients
The symbolic gradients of the shape functions with respect to $\xi$ are:


```python
# Compute gradients
dN_dxi = [diff(Ni, xi) for Ni in N]

# Print gradients
for i in range(2):
    print(f"{dN_dxi[i]};")
```

## Numerical Evaluation
We can specialize these general expressions for specific reference elements.

### Unit Interval [0, 1]


```python
# Set numerical values for reference element (e.g., unit interval)
xi0_val, xi1_val = 0, 1
```

```python
# Substitute numerical values into shape functions
N_numeric = [Ni.subs({nodes[0]: xi0_val, nodes[1]: xi1_val}) for Ni in N]
dN_dxi_numeric = [dNi.subs({nodes[0]: xi0_val, nodes[1]: xi1_val}) for dNi in dN_dxi]
```

```python
# Print results with numerical values
print(f"(xi0={xi0_val}, xi1={xi1_val})")

print("\nShape functions:")
for i in range(2):
    print(f"{N_numeric[i]};")

print("\nGradients:")
for i in range(2):
    print(f"{dN_dxi_numeric[i]};")
```

### Standard Reference Interval [-1, 1]


```python
xi0_val2, xi1_val2 = -1, 1
N_numeric2 = [Ni.subs({nodes[0]: xi0_val2, nodes[1]: xi1_val2}) for Ni in N]
dN_dxi_numeric2 = [dNi.subs({nodes[0]: xi0_val2, nodes[1]: xi1_val2}) for dNi in dN_dxi]

# Print results with numerical values
print(f"(xi0={xi0_val2}, xi1={xi1_val2})")

print("\nShape functions:")
for i in range(2):
    print(f"{N_numeric2[i]};")

print("\nGradients:")
for i in range(2):
    print(f"{dN_dxi_numeric2[i]};")
```
