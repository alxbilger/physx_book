# Shape Functions on linear triangle

This notebook derives the linear shape functions for a 2D finite element (triangle).
Shape functions $N_i$ are used to interpolate values within an element based on nodal values.


```python
from sympy import symbols, Eq, solve, simplify, IndexedBase, diff, factor
```

```python
from typst_utils import typst_print
```

## Symbol Definition
We define the 2D position vector $\mathbf{X} = [\xi, \eta]^T$ and the positions of the three nodes $\mathbf{X}_i$ for $i=0, 1, 2$.


```python
# Define symbols
X = IndexedBase('referenceposition')
# Using IndexedBase for each node to represent vectors
nodes = [IndexedBase(f'referenceposition^(({i}))') for i in range(3)]
```

## Linear Combination
We assume that each shape function $N_i$ is a linear function of the components of $\mathbf{X}$:
$N_i(\mathbf{X}) = a \cdot X[0] + b \cdot X[1] + c$


```python
# Define shape functions as linear combinations: Ni = a * X[0] + b * X[1] + c
a, b, c = symbols('a b c')
```

## Solving the System
The shape functions must satisfy the Kronecker delta property: $N_i(\mathbf{X}_j) = \delta_{ij}$.
This leads to a system of three linear equations for the coefficients $a, b$ and $c$ for each $N_i$.


```python
# Solve for each shape function N[i]
N = []
for i in range(3):
    # Ni(X) = a*X[0] + b*X[1] + c
    eqs = [
        Eq(a * nodes[j][0] + b * nodes[j][1] + c, 1 if i == j else 0)
        for j in range(3)
    ]
    sol = solve(eqs, (a, b, c))
    print(f"for N[{i}]:\na={sol[a].simplify()};\nb={sol[b].simplify()};\nc={sol[c].simplify()}\n\n")
    Ni = sol[a] * X[0] + sol[b] * X[1] + sol[c]
    N.append(factor(Ni))
```

## Symbolic Results
The general symbolic expressions for the shape functions are:


```python
# Print results
print("Shape functions:")
for i in range(3):
    print(f"{N[i]};\n")
```

## Symbolic Gradients
The symbolic gradients of the shape functions with respect to $\xi$ and $\eta$ are:


```python
# Compute gradients
dN_dxi = [diff(Ni, X[0]) for Ni in N]
dN_deta = [diff(Ni, X[1]) for Ni in N]

# Print gradients
for i in range(3):
    print(f"Gradient dN[{i}]/dxi: {typst_print(dN_dxi[i])}")
    print(f"Gradient dN[{i}]/deta: {typst_print(dN_deta[i])}")
    print()
```

## Numerical Evaluation
We can specialize these general expressions for specific reference elements.


### Standard Reference Triangle (0,0), (1,0), (0,1)


```python
# Standard reference triangle values
vals_std = {
    nodes[0][0]: 0, nodes[0][1]: 0,
    nodes[1][0]: 1, nodes[1][1]: 0,
    nodes[2][0]: 0, nodes[2][1]: 1
}

N_numeric_std = [Ni.subs(vals_std) for Ni in N]
dN_dxi_numeric_std = [dNi.subs(vals_std) for dNi in dN_dxi]
dN_deta_numeric_std = [dNi.subs(vals_std) for dNi in dN_deta]

for i in range(3):
    print(f"Shape function N[{i}] (standard): {N_numeric_std[i]}")
    print(f"Gradient dN[{i}]/dxi (standard): {dN_dxi_numeric_std[i]}")
    print(f"Gradient dN[{i}]/deta (standard): {dN_deta_numeric_std[i]}")
    print()
```
