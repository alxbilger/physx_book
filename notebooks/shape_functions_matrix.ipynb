```python
from sympy import symbols, Matrix, simplify, latex
from typst_utils import typst_print
```

```python
# Define symbols for shape functions
N0, N1, N2, N3 = symbols('N_0 N_1 N_2 N_3')

# Define the matrix N (3x12)
N = Matrix([
    [N0, 0, 0, N1, 0, 0, N2, 0, 0, N3, 0, 0],
    [0, N0, 0, 0, N1, 0, 0, N2, 0, 0, N3, 0],
    [0, 0, N0, 0, 0, N1, 0, 0, N2, 0, 0, N3]
])

# Compute N^T * N
NTN = N.T * N

# Print the result
print("Matrix N:")
print(typst_print(N))

print("\nMatrix N^T * N:")
print(typst_print(NTN))
```
